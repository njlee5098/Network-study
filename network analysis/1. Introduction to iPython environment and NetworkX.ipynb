```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Zachary's Karate Club

```python
G = nx.read_gml('./data/karate.gml', label='id')
nx.draw_networkx(G)
```

```python
k = dict(nx.degree(G))
plt.figure(figsize=(10,7))
coord = nx.spring_layout(G)

nx.draw_networkx(G,
                 pos=coord,
                 nodelist=k.keys(),
                 node_size = [d*100 for d in k.values()],
                 node_color=list(k.values()),
                 font_size=8,
                 cmap=plt.cm.Reds,
                 )
```

```python
# load adjacency matrix
A = nx.to_numpy_matrix(G, dtype=np.bool)
k = dict(G.degree())


# sort nodes according to degree
k_sorted = dict(sorted(k.items(), key=lambda item: item[1], reverse=True))
AA = nx.to_numpy_matrix(G, dtype=np.bool, nodelist=k_sorted.keys())

# show adjacency matrix

plt.figure(figsize=(20,10))

plt.subplot(121)
plt.imshow(A,
           cmap="Greys",
           interpolation="none"
           )
plt.title('Zachary\'s club adjacency matrix', fontsize=14)

# Show sorted adjacency matrix
plt.subplot(122)
plt.imshow(AA,
           cmap="Greys",
           interpolation="none"
           )
plt.title('Zachary\'s club adjacency matrix - sorted by degree', fontsize=14)
plt.figure()
```

```python
# show node degrees
plt.figure(1, figsize=(10, 5))
plt.bar(np.arange(34), k_sorted.values())
plt.xticks(np.arange(34), k_sorted.keys())
plt.margins(0.2)
plt.ylabel('Degree')
plt.xlabel('Node id')
```
